```python
import math
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
import statistics
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/MTECH_PROJECT/s10/s10_emg.csv')
print(df1.shape)
df1.head()
```

```python
df2 = pd.read_csv('/content/drive/MyDrive/MTECH_PROJECT/s10/s10_Kinematics.csv')
print(df2.shape)
df2.head()
```

```python
#####nan value check
df1.dropna(inplace=True)
df2.dropna(inplace=True)
```

```python
#####input data partition
X_train, X_test, y_train, y_test = train_test_split(df1, df2, train_size=0.70, random_state=42)

print(f' Training input data size: {X_train.shape}')
print(f' Test input data size: {X_test.shape}')
print(f' Training output data size: {y_train.shape}')
print(f' Test output data size: {y_test.shape}')
```

```python
#####Linear regression
lr = LinearRegression()
start1=time.time()
lr.fit(X_train,y_train)
end1=time.time()
lr_pred_train = lr.predict(X_train)
start2=time.time()
lr_pred_test = lr.predict(X_test)
end2=time.time()
lr_accuracy = pd.DataFrame({
    'Parameters': ['R2_train','RMSE_train','R2_test','RMSE_test','time taken1','time taken2'],
    'lr': [ r2_score(y_train,lr_pred_train),mean_squared_error(y_train,lr_pred_train,squared=False), 
                 r2_score(y_test,lr_pred_test),mean_squared_error(y_test,lr_pred_test,squared=False),(end1-start1)/98000,(end2-start2)/42000]})
lr_accuracy=lr_accuracy.transpose()
print(lr_accuracy)
```
