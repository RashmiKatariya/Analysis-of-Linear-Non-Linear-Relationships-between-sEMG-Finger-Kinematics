```python
import math
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.neighbors import KNeighborsRegressor
import statistics
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/MTECH_PROJECT/s9/emg_random_s9.csv')
print(df1.shape)
df1.head()
```

```python
df2 = pd.read_csv('/content/drive/MyDrive/MTECH_PROJECT/s9/kinematics_random_s9.csv')
print(df2.shape)
df2.head()
```

```python
#####nan value check
df1.dropna(inplace=True)
df2.dropna(inplace=True)
```

```python
#####input data partition
X_train, X_test, y_train, y_test = train_test_split(df1, df2, train_size=0.70, random_state=42)

print(f' Training input data size: {X_train.shape}')
print(f' Test input data size: {X_test.shape}')
print(f' Training output data size: {y_train.shape}')
print(f' Test output data size: {y_test.shape}')
```

```python
#####KNN regression
knn = KNeighborsRegressor()
start1=time.time()
knn.fit(X_train,y_train)
end1=time.time()
# print("time taken1: ",(end1-start1)/98000)
knn_pred_train = knn.predict(X_train)
start2=time.time()
knn_pred_test = knn.predict(X_test)
end2=time.time()
# print("time taken2: ",(end2-start2)/42000)
knn_accuracy = pd.DataFrame({
    'Parameters': ['R2_train','RMSE_train','R2_test','RMSE_test','time taken1','time taken2'],
    'knn': [ r2_score(y_train,knn_pred_train),mean_squared_error(y_train,knn_pred_train,squared=False), 
                 r2_score(y_test,knn_pred_test),mean_squared_error(y_test,knn_pred_test,squared=False),(end1-start1)/98000,(end2-start2)/42000]})
knn_accuracy=knn_accuracy.transpose()
print(knn_accuracy)
# print("time taken: ",end-start)
```
