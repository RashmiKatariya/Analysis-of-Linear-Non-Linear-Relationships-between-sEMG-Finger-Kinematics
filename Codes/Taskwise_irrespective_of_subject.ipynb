```python
import math
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
import statistics
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/MTECH_PROJECT/Thumb/final_thumb_emg.csv')
print(df1.shape)
df1.head()
```

```python
df2 = pd.read_csv('/content/drive/MyDrive/MTECH_PROJECT/Thumb/final_thumb_kinematics.csv')
print(df2.shape)
df2.head()
```

```python
#####input data partition
X_train, X_test, y_train, y_test = train_test_split(df1, df2, train_size=0.70, random_state=42)

print(f' Training input data size: {X_train.shape}')
print(f' Test input data size: {X_test.shape}')
print(f' Training output data size: {y_train.shape}')
print(f' Test output data size: {y_test.shape}')
```

```python
#####Random Forest regression
regressor = RandomForestRegressor(n_estimators=100,random_state=0)
start1=time.time()
regressor.fit(X_train,y_train)
end1=time.time()
regressor_pred_train = regressor.predict(X_train)
start2=time.time()
regressor_pred_test = regressor.predict(X_test)
end2=time.time()
regressor_accuracy = pd.DataFrame({
    'Parameters': ['R2_train','RMSE_train','R2_test','RMSE_test','time taken1','time taken2'],
    'regressor': [ r2_score(y_train,regressor_pred_train),mean_squared_error(y_train,regressor_pred_train,squared=False), 
                 r2_score(y_test,regressor_pred_test),mean_squared_error(y_test,regressor_pred_test,squared=False),(end1-start1)/98000,(end2-start2)/42000]})
regressor_accuracy=regressor_accuracy.transpose()
print(regressor_accuracy)
```
