```python
import numpy as np
import matplotlib.pyplot as plt

# Create example data
subjects = ['S1', 'S2', 'S3', 'S4', 'S5','S6', 'S7', 'S8', 'S9', 'S10']
rmse_scores = np.array([0.501822,
0.692507,
0.858421,
0.989112,
0.654958,
0.815048,
0.55325,
0.5386,
1.179228,
0.621364

])
r2_scores = np.array([0.994261,
0.994083,
0.994641,
0.989112,
0.996476,
0.992855,
0.995741,
0.997691,
0.983364,
0.99466

])

# Set the width of the bars
bar_width = 0.35

# Set the position of the bars on the x-axis
r1 = np.arange(len(subjects))
r2 = [x + bar_width for x in r1]

# Plot the bars
plt.bar(r1, rmse_scores, color='blue', width=bar_width, edgecolor='grey', label='RMSE')
plt.bar(r2, r2_scores, color='orange', width=bar_width, edgecolor='grey', label='R2 score')

# Add xticks on the middle of the group bars
plt.title('Random Forest Regressor')
plt.xlabel('Subjects', fontweight='bold')
plt.ylabel('Scores', fontweight='bold')
plt.xticks([r + bar_width/2 for r in range(len(subjects))], subjects)

# Add legend
plt.legend()

# Show the plot
plt.show()
```
